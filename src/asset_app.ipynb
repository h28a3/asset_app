```python
import time
import chromedriver_binary
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome import service

#WEBブラウザの起動
driver = webdriver.Chrome() 

#自動ログインしたいウェブサイトのURLをコピペ
driver.get('https://site1.sbisec.co.jp/ETGate/?_ControlID=WPLEThmR001Control&_PageID=DefaultPID&_DataStoreID=DSWPLEThmR001Control&_ActionID=DefaultAID&getFlg=on') 

#例を参照して、属性name,classの値を入力
my_id = driver.find_element(By.NAME, "user_id")
password = driver.find_element(By.NAME, "user_password")
submit = driver.find_element(By.NAME, "ACT_login")

#入力欄を空にする
my_id.clear()
password.clear()

#自動入力したIDとパスワードを入力
my_id.send_keys("idを入力")
password.send_keys("パスワードを入力")

#ログインボタンを自動入力
submit.click()

# 口座管理に遷移
driver.find_element(By.XPATH, '//*[@id="link02M"]/ul/li[3]/a/img').click()

# 円建ての資産総額を取得
sbi = float(driver.find_element(By.XPATH, '/html/body/div[1]/table/tbody/tr/td[1]/table/tbody/tr[2]/td/table[1]/tbody/tr/td/form/table[2]/tbody/tr[1]/td[2]/table[4]/tbody/tr/td[1]/table[5]/tbody/tr[8]/td[2]/div/b').text.replace(',',''))

driver.get('https://member.rakuten-sec.co.jp/app/com_page_template.do;BV_SessionID=F69B54129B16A541E44DDB215C93E082.79312000?BV_SessionID=F69B54129B16A541E44DDB215C93E082.79312000') 

my_id = driver.find_element(By.NAME, "loginid")
password = driver.find_element(By.NAME, "passwd")
submit = driver.find_element(By.NAME, "submit%template")

#入力欄を空にする
my_id.clear()
password.clear()

#自動入力したIDとパスワードを入力
my_id.send_keys("idを入力")
password.send_keys("パスワードを入力")

#ログインボタンを自動入力
submit.click()

rakuten_bank = float(driver.find_element(By.XPATH, '//*[@id="asset_total_power_amount"]').text.replace(',',''))
driver.find_element(By.XPATH, '//*[@id="homeAssetsTrigger"]').click()
spot = float(driver.find_element(By.XPATH, '//*[@id="balance_data_actual_data"]/section/table/tbody/tr[1]/td[1]/p/span[1]').text.replace(',',''))
time.sleep(5)
rakuten = float(driver.find_element(By.XPATH, '//*[@id="balance_data_actual_data"]/section/table/tbody/tr[6]/td[1]/p/span[1]').text.replace(',',''))

driver.get('https://www.netbk.co.jp/contents/pages/wpl010101/i010101CT/DI01010210') 

my_id = driver.find_element(By.NAME, "userNameNewLogin")
password = driver.find_element(By.ID, "loginPwdSet")
submit = driver.find_element(By.XPATH, "/html/body/app/div[1]/ng-component/div/main/ng-component/form/section[1]/div/div/ul/li/nb-button-login/button/span")

#入力欄を空にする
my_id.clear()
password.clear()

#自動入力したIDとパスワードを入力
my_id.send_keys("idを入力")
password.send_keys("パスワードを入力")

#ログインボタンを自動入力
submit.click()
time.sleep(5)

sbi_bank = float(driver.find_element(By.XPATH, '/html/body/app/div[1]/ng-component/div/main/ng-component/div[1]/div/nb-titlecarousel/div/div/ul/li/div/div/strong').text.replace(',',''))

driver.get('https://trade.matsui.co.jp/mgap/login') 

my_id = driver.find_element(By.NAME, "user")
password = driver.find_element(By.NAME, "password")
submit = driver.find_element(By.XPATH, '//*[@id="login-form"]/main/div[2]/div[4]/div/div/div[2]/button')

#入力欄を空にする
my_id.clear()
password.clear()

#自動入力したIDとパスワードを入力
my_id.send_keys("idを入力")
password.send_keys("パスワードを入力")

#ログインボタンを自動入力
submit.click()

time.sleep(10)
matsui_bank = driver.find_element(By.XPATH, '//*[@id="top-page"]/div[2]/div[1]/div/div[1]/div[2]/div[1]/div[4]/div[4]').text.replace(',','')
matsui_bank = float(matsui_bank.replace('円',''))

driver.get('https://direct.jp-bank.japanpost.jp/tp1web/U010101WAK.do?link_id=ycDctLgn') 

time.sleep(1)
my_id = driver.find_element(By.NAME, "okyakusamaBangou1")
#my_id.clear()
my_id.send_keys("idを入力(13桁)")
my_id = driver.find_element(By.NAME, "okyakusamaBangou2")
my_id.send_keys("idを入力(4桁)")
my_id = driver.find_element(By.NAME, "okyakusamaBangou3")
my_id.send_keys("idを入力(5桁)")
submit = driver.find_element(By.NAME, 'U010103')
submit.click()

time.sleep(10)
password = driver.find_element(By.NAME, "loginPassword")
submit = driver.find_element(By.NAME, 'U010302')
#password.clear()
password.send_keys("パスワードを入力")

#ログインボタンを自動入力
submit.click()

time.sleep(4)
post_bank = float(driver.find_element(By.XPATH, '//*[@id="strMain"]/div[1]/div/div[2]/p/span').text.replace(',',''))

# 閉じる
driver.close()
cash = int(sbi_bank + rakuten_bank + matsui_bank + post_bank)
trust = int(rakuten + sbi)
spot = int(spot)
```

```python
import numpy as np
import matplotlib.pyplot as plt

asset = np.array([cash,trust,spot])
asset_nm = np.array(["cash", "trust", "spot"])
pie_colors = ["r", "c", "b"]
plt.title("asset",fontsize = 22)
plt.rcParams['font.size'] = 16.0 #ラベルのフォントサイズ

plt.pie(asset, labels=asset_nm, counterclock=True,autopct="%1.1f%%", colors=pie_colors) # (6)円グラフの描画
plt.show()
```

```python
import openpyxl as op
from openpyxl.styles.alignment import Alignment
import datetime

today = datetime.datetime.now().strftime('%Y/%m')
wb = op.load_workbook('asset.xlsx')
ws = wb["Sheet1"]
max_row = ws.max_row + 1
ws.cell(row = max_row,column = 1).value = today
ws.cell(row = max_row,column = 2).value = cash
ws.cell(row = max_row,column = 3).value = trust
ws.cell(row = max_row,column = 4).value = spot
ws.cell(row = max_row,column = 5).value = cash + trust + spot
ws.cell(row = max_row,column = 1).alignment = Alignment(horizontal = 'right', vertical = 'center')
wb.save('asset.xlsx')
```
